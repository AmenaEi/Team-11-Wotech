```python
# Sample data representing the average temperature (in degrees Celsius) from 2010 to 2014
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Avg_Temperature': [14.1, 14.3, 14.5, 14.2, 14.6]}
```

```python
import pandas as pd
import matplotlib.pyplot as plt #commonly used abbriviations
```

```python
df = pd.DataFrame(data) #create a dataframe
df
```

```python
# Version one - when want a create a spicific customization for a plot - this one is slightly more complicated

plt.plot(df['Year'], df['Avg_Temperature'], marker = 'o') #df[name of the column] - how to refer to datframe's column
plt.title('Average Temperature from 2010 to 2014')
plt.xlabel('Year') #title on x axix
plt.ylabel('Temperature')  #title on y axis

plt.xticks(df['Year']) #x axis ticks - fixed x axis, are now rounded, without 2010.5
plt.grid(True) #add a grid
plt.show() #hides the logs of visualizations
```

```python
# Version two! - high level plotting function - can quickly show visualization

df.plot(x='Year', y='Avg_Temperature', kind='line', title='Average Temperature from 2010 to 2014', marker = 'o')
plt.xlabel('Year')
plt.ylabel('Temperature')

plt.xticks(df['Year']) #x axis ticks - fixed x axis, are now rounded, without 2010.5
plt.grid(True) #add a grid
plt.show()
```

```python
# BAR CHART
plt.bar(df['Year'], df['Avg_Temperature'], color='darkred') #color can be selected by you
plt.title('Average Temperature from 2010 to 2014')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.show()
```

```python
# Horizontal bar chart

plt.barh(df['Year'], df['Avg_Temperature'], color='navy') # barH !
plt.title('Average Temperature from 2010 to 2014')
plt.xlabel('Temperature')
plt.ylabel('Year')
plt.show()
```

```python
#Histogram - presenta the distribution of a single variable, frequency, represents numerical data

data2 = {'Scores': [85, 92, 78, 20, 89, 90, 56, 60, 23, 90, 40, 77]}
df2 = pd.DataFrame(data2)
df2
```

```python
plt.hist(df2['Scores'], bins=3, color='darkorange', edgecolor='black')
plt.title('Histogram of Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')

plt.show()

#scores from 20 to 43 are 3, there are 2 scores from 43 to 68, acores above 68 - 7 students
#change bins, see different distributions (how many values of each group)
```

```python
bin_edges = [0, 40, 60, 80, 90, 100]  #Create histogram with custom bin edges
plt.hist(df2['Scores'], bins=bin_edges, color='darkorange', edgecolor='black')
plt.title('Histogram of Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')

plt.show()
```

```python
# Scatter plot - used to visualize the relationship two variables
data3 = {'X': [1,2,3,5,5,2], 'Y':[5,8,7,4,9,5]}

df3 = pd.DataFrame(data3)
df3
plt.scatter(x = df3['X'], y=df3['Y'])
plt.title('Scatter Plot')
plt.grid(True)
plt.show()
```

```python
# PIE CHART - represents parts of a whole

data4 = {'Category': ['A', 'B', 'C', 'D'], 'Values': [15, 30, 25, 40]}
df4 = pd.DataFrame(data4)
df4
```

```python
plt.pie(df4['Values'], labels = df4['Category'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
df4
```

```python
# Box plot - distribution of a dataset and indentifies outliers

plt.boxplot(df4['Values'])
plt.title('Box Plot')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('/content/transaction_dataset.csv')
df = pd.DataFrame(dataset)

# Count occurrences of each unique value in 'Gender' including NaN values
gender_counts = df['Gender'].value_counts(dropna=False)

# Create the bar chart
gender_counts.plot(kind='bar')

# Customize the plot
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Customers by Gender (including NaN values)')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
dataset2 = pd.read_csv('/content/transaction_dataset.csv')

# Count occurrences of each unique value in 'Gender' column (including NaN)

# fillna('NaN') method replaces NaN values with the string 'NaN' before counting the occurrences.
# This ensures that the gender_counts2 series contains only strings,
# which can be used to plot the bar chart without encountering a TypeError.
gender_counts2 = dataset2['Gender'].fillna('NaN').value_counts()

# Prepare data for plotting
genders = gender_counts2.index.tolist()
counts = gender_counts2.values

# Plot bar chart
plt.bar(genders, counts, color='red')
plt.title('Count of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
