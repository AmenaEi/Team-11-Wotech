###Data import

```python
import seaborn as sns
import pandas as pd

titanic_data = sns.load_dataset('titanic')
```

```python
titanic_data.head()
```

###Data investigation

```python
titanic_data.isnull().sum()  #how many values is NULL (True-False)
```

```python
titanic_data.dtypes  #data types
```

```python
titanic_data.drop('deck', axis=1, inplace=True)  #delete column with a lot of NULL values
```

###Data Transformation

```python
titanic_data['class'] = titanic_data['class'].astype(str) #modified the object from category to string

titanic_data.fillna(0, inplace=True) #replaces empty values with 0
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.drop(['alive', 'alone', 'adult_male'], axis=1, inplace=True)  #drop 3 columns
```

```python
titanic_data.head()
```

```python
titanic_data.drop('class', axis=1, inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data['sex'].unique()  #unique values
```

```python
titanic_data['who'].unique()
```

```python
# Sex column

sex_mapping = {'male': 0, 'female': 1}
titanic_data['sex'] = titanic_data['sex'].map(sex_mapping) #replace sex column and map each value in this column to the mapping
```

```python
titanic_data.head()
```

```python
# Embark town column

titanic_data['embark_town'].unique() # 0 - null values that were replaced
```

```python
embark_town_mapping = {'Southampton': 1, 'Cherbourg': 2, 'Queenstown': 3, 0:0}
titanic_data['embark_town'] = titanic_data['embark_town'].map(embark_town_mapping)
titanic_data.head()
```

```python
# Embarked column

titanic_data['embarked'].unique()
```

```python
embarked_mapping = {'S': 1, 'C': 2, 'Q': 3, 0:0}
titanic_data['embarked'] = titanic_data['embarked'].map(embarked_mapping)
titanic_data.head()
```

```python
# Who column

titanic_data['who'].unique()
```

```python
who_mapping = {'man': 1, 'woman': 2, 'child': 3}
titanic_data['who'] = titanic_data['who'].map(who_mapping)
titanic_data.head()

#it also can be done differently - label encoder
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html
```

```python
titanic_data.info()
```

###Data modeling

```python
from sklearn.model_selection import train_test_split
X = titanic_data.drop('survived', axis=1) # FEATURES - whole data frame except 'survived' column
y = titanic_data['survived'] #survived column separately
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
#we need the random state so that if we run this line of code we get same results, the split will be the same
```

###Logistic regression

```python
from sklearn.linear_model import LogisticRegression #import
model = LogisticRegression() #initialize
model.fit(X_train, y_train) #train
```

```python
prediction_results = model.predict(X_test) #test
prediction_results  #0 - did not survive, 1 - person survived
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)
predict_tree = tree_clf.predict(X_test)
predict_tree
```

1. tree_clf = DecisionTreeClassifier(random_state=42)

* What's happening: This line is creating a new "Decision Tree Classifier" model. Think of it like building a decision-making tree that helps you decide whether someone survives or not based on different factors (like age, ticket class, etc.).

* random_state=42: This is like setting a specific seed so that every time you build this tree, it grows in the same way. It's a bit like planting the same type of tree in the same spot every time—it’ll always look the same.

2. tree_clf.fit(X_train, y_train)

* What's happening: Here, the tree is being trained using the data. The X_train is the data with all the information about the passengers (like age, class, etc.), and y_train is the answer key (whether they survived or not). The tree uses this information to learn patterns and rules to make decisions.

* fit: This means "train the model." The tree is learning how to make decisions based on the training data.

3. predict_tree = tree_clf.predict(X_test)

* What's happening: Now that the tree has learned how to make decisions, this line uses the tree to predict what happens to a new set of passengers (X_test) who weren't part of the training. The tree looks at each passenger's information and decides whether they survived or not.
* predict: This means "make predictions" using the tree that was just trained.

4. predict_tree

What's happening: Finally, this stores the predictions made by the tree for the new passengers in predict_tree. You could print this or check how many the tree got right or wrong by comparing it with the real outcomes.

The result you see is an array of predictions made by the Decision Tree model on the test data (X_test). Each number in the array represents the model's prediction for whether a passenger survived or not:

1: This means the model predicts that the passenger survived.

0: This means the model predicts that the passenger did not survive.

For example, if the array starts with [1, 0, 1, 1, 0, ...], it means:

* The first passenger in the test data is predicted to have survived (1).
* The second passenger is predicted not to have survived (0).
* The third passenger is predicted to have survived (1).

And so on.


This array is useful because you can compare these predictions to the actual outcomes (whether each passenger really survived or not) to see how accurate the model is.
