```python
import numpy as np
```

```python
coin = ['Heads', 'Tails'] # 1-dimensional array
```

```python
coin
```

```python
number_of_throws = 1000 #size of the random list that we want to generate
```

```python
'''
random.choice(a, size=None, replace=True, p=None)
a
1-D array-like or int
  If an ndarray, a random sample is generated from its elements. If an int, the random sample is generated as if it were np.arange(a)

size
int or tuple of ints, optional
  Output shape. If the given shape is, e.g., (m, n, k), then m * n * k samples are drawn. Default is None, in which case a single value is returned.
'''

tosses = np.random.choice(coin, size=number_of_throws)
```

```python
print(f"The results of coin toss simulation for {number_of_throws} throws are: \n{tosses}")
```

```python
# Numbers of Heads
total_head_tosses = np.sum(tosses == 'Heads') # count the number of heads
print(f'Total head count is {total_head_tosses}')
```

```python
# Number of Tails
total_tail_tosses = np.sum(tosses == 'Tails') # count the number of heads
print(f'Total tail count is {total_tail_tosses}')
```

```python
#Probabilities
prob_heads = total_head_tosses/number_of_throws
prob_tails = total_tail_tosses/number_of_throws

print(f'Probability of getting heads is {prob_heads}')
print(f'Probability of getting tails is {prob_tails}')
```

###Group task

Design a simulation where you toss two coins simultaneously. Record the possible outcomes (heads or tails for each coin) and run the simulation for multiple trials (e.g., 100, 500, or 1000 tosses). After collecting the data:

* Calculate the probabilities of each outcome (HH, HT, TH, TT).
* Visualize the results using a bar chart or pie chart to represent the frequencies and probabilities of each outcome.
* Analyze whether the results align with the expected theoretical probabilities.

```python
import numpy as np
import matplotlib.pyplot as plt

# Possible outcomes for two coin tosses
outcomes = ['HH', 'HT', 'TH', 'TT']

# Number of trials (coin tosses)
number_of_tosses = 1000

# Simulate tossing two coins simultaneously for the given number of trials
coin1 = np.random.choice(['Heads', 'Tails'], size=number_of_tosses)
coin2 = np.random.choice(['Heads', 'Tails'], size=number_of_tosses)

# Combine the results of the two coin tosses to get the outcomes (HH, HT, TH, TT)
combined_tosses = [coin1[i] + coin2[i] for i in range(number_of_tosses)]

# Map the combined tosses to the correct format for outcomes (e.g., 'HeadsHeads' to 'HH')
combined_tosses = ['HH' if x == 'HeadsHeads' else 'HT' if x == 'HeadsTails' else 'TH' if x == 'TailsHeads' else 'TT' for x in combined_tosses]

# Count the occurrences of each outcome
outcome_counts = {
    'HH': np.sum(np.array(combined_tosses) == 'HH'),
    'HT': np.sum(np.array(combined_tosses) == 'HT'),
    'TH': np.sum(np.array(combined_tosses) == 'TH'),
    'TT': np.sum(np.array(combined_tosses) == 'TT')
}

# Calculate the probabilities of each outcome
total_tosses = number_of_tosses
probabilities = {outcome: count / total_tosses for outcome, count in outcome_counts.items()}

# Print the counts and probabilities
print(f"Outcome counts: {outcome_counts}")
print(f"Probabilities: {probabilities}")

# Visualization: Bar chart of frequencies
plt.bar(outcome_counts.keys(), outcome_counts.values(), color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Outcomes (HH, HT, TH, TT)')
plt.ylabel('Frequencies')
plt.title(f'Frequencies of Two Coin Tosses ({number_of_tosses} tosses)')
plt.show()

print()

# Visualization: Pie chart of probabilities
plt.pie(probabilities.values(), labels=probabilities.keys(), autopct='%1.1f%%', colors=['blue', 'orange', 'green', 'red'])
plt.title(f'Probabilities of Two Coin Tosses ({number_of_tosses} tosses)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Possible outcomes for two coin tosses
outcomes = ['HH', 'HT', 'TH', 'TT']

# Number of trials (coin tosses)
number_of_tosses = 1000

# Simulate tossing two coins simultaneously for the given number of trials
coin1 = np.random.choice(['H', 'T'], size=number_of_tosses)
coin2 = np.random.choice(['H', 'T'], size=number_of_tosses)

# Combine the results of the two coin tosses to get the outcomes (HH, HT, TH, TT)
combined_tosses = np.char.add(coin1, coin2)  # directly combine coin1 and coin2

# Count the occurrences of each outcome
outcome_counts = {
    'HH': np.sum(combined_tosses == 'HH'),
    'HT': np.sum(combined_tosses == 'HT'),
    'TH': np.sum(combined_tosses == 'TH'),
    'TT': np.sum(combined_tosses == 'TT')
}

# Calculate the probabilities of each outcome
total_tosses = number_of_tosses
probabilities = {outcome: count / total_tosses for outcome, count in outcome_counts.items()}

# Print the counts and probabilities
print(f"Outcome counts: {outcome_counts}")
print(f"Probabilities: {probabilities}")

print()

# Visualization: Bar chart of frequencies
plt.bar(outcome_counts.keys(), outcome_counts.values(), color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Outcomes (HH, HT, TH, TT)')
plt.ylabel('Frequencies')
plt.title(f'Frequencies of Two Coin Tosses ({number_of_tosses} tosses)')
plt.show()

print()

# Visualization: Pie chart of probabilities
plt.pie(probabilities.values(), labels=probabilities.keys(), autopct='%1.1f%%', colors=['blue', 'orange', 'green', 'red'])
plt.title(f'Probabilities of Two Coin Tosses ({number_of_tosses} tosses)')
plt.show()
```
