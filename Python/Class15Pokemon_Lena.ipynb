```python
#required libraries

import pandas as pd #analysis
import matplotlib.pyplot as plt #visualization
```

```python
pokemon_df = pd.read_csv('/content/Pokemon.csv')
pokemon_df.head(n=10) #head will give you top 10 rows
```

```python
# How many pokemons have Defense greater than 100?

defense_greater_than_100 = pokemon_df[pokemon_df['Defense'] > 100]
print(len(defense_greater_than_100))
```

```python
#Which pokemon has the biggest attack?

pokemon_df['Attack'].max() #the hightst value of the attack
pokemon_df['Attack'].idxmax() #index of the highest value

#pokemon_df[pokemon_df['Attack'] == pokemon_df['Attack'].max()]    #returns the line in the table
highest_attack = pokemon_df['Name'].loc[pokemon_df['Attack'].idxmax()]
print(f'The pokemon with the highest attack is {highest_attack}')
```

```python
#Which pokemon has the smallest attack?

pokemon_df['Attack'].min() #the hightst value of the attack
pokemon_df['Attack'].idxmin() #index of the highest value

lowest_attack = pokemon_df['Name'].loc[pokemon_df['Attack'].idxmin()]
print(f'The pokemon with the highest attack is {lowest_attack}')
```

### HISTOGRAM

```python
#analyze total for each pokemon
plt.hist(pokemon_df['Total'], color='maroon', edgecolor='black', alpha=0.7) #alpha = transparency, default = 10bins
plt.xlabel('Total')
plt.ylabel('Number of pokemons (Frequency)')
plt.title('Distribution of Total Values')

#average line
#axvline - draw a line on X axis Vertically
#pokemon_df['Total'].mean() = average
plt.axvline(pokemon_df['Total'].mean(), color='red', linestyle='dashed')

#min line
plt.axvline(pokemon_df['Total'].min(), color='lightgrey', linestyle='dashed')
#max line
plt.axvline(pokemon_df['Total'].max(), color='grey', linestyle='dashed')

plt.grid(linestyle='--', alpha=0.7, color='pink')

'''
To add the labels, including the avg line:
# Annotate the histogram bins
for i in range(len(bins) - 1):
    plt.text(bins[i] + (bins[i+1] - bins[i]) / 2, n[i], int(n[i]), ha='center', va='bottom')

# Label the average line
avg = pokemon_df['Total'].mean()
plt.text(avg, plt.ylim()[1] * 0.9, f'Avg: {avg:.2f}', color='maroon', ha='center', rotation=90)
'''

plt.show()
```

```python
value = plt.hist(pokemon_df['Total'], bins = 3, color='darkblue', edgecolor='black', alpha=0.7)
hist_values = value[0]
bin_edges = value[1]
#what does [0] and [1]
#accessing element 0 or 1 from the object that is created by hist

print('Histogram counts:', hist_values) #how many values are in each bars
print('Bin edges', bin_edges) #where the bar starts, and where ends (the end value is not included)

plt.show()
```

```python
'''
1.pokemon_df['Total']: This is like asking for a list of the "Total" scores for
all the Pokémon.
579: This is checking each Pokémon to see if their "Total" score is more than 579.
(pokemon_df['Total'] > 579): This makes a new list where each Pokémon gets
a "yes" or "no" (True or False) answer depending on if their score is more than 579.
.sum(): This adds up all the "yes" answers. Since "yes" (True) counts as 1 and "no"
(False) counts as 0, it gives us the total number of Pokémon that have a score
higher than 579.
'''

(pokemon_df['Total']>579).sum()
#filtering the total above 579 and summing them up
#it computes the number of Pokémon in the pokemon_df DataFrame that have a 'Total' stat greater than 579.
```

```python
#How many Pokémons are with 'Type 1' == 'Water' as a percent of total

total_num_of_pokemons = pokemon_df['Type 1'].value_counts().sum()
'''
value_counts() function is applied to a column and it returns a Series where
the index is the unique values from that column and the values are the counts of occurrences of each unique value.
For example, if the 'Type 1' column contains multiple Pokémon with types 'Water', 'Fire', and 'Grass',
value_counts() will count how many Pokémon there are of each type
Alternatives?:
total_num_of_pokemons = pokemon_df.shape[0]
total_num_of_pokemons = pokemon_df['Type 1'].count()

'''
water_pokemons = pokemon_df['Type 1'].value_counts()['Water']
#Directly retrieves the count of 'Water' type Pokémon from a summary of all types.
#Returns an integer count of how many Pokémon are of type 'Water'.
#Counts how many times 'Water' appears in the 'Type 1' column.
#It counts occurrences of each unique value in a Series

'''
#alternative
water_type = pokemon_df['Type 1'] == 'Water'
Creates a boolean mask indicating which rows are of type 'Water'.
Returns a boolean Series with True where the type is 'Water' and False elsewhere.
'''
water_percent = water_pokemons / total_num_of_pokemons * 100
print(f'The percent of water pokemons is {round(water_percent, 1)}%')
```

```python
#Filter the DataFrame to include only the Pokémon with 'Speed' >= 80

pokemon_df[pokemon_df['Speed'] >= 80]
```

```python
#How many Pokémon meet this criterion?

'''
pokemon_df[pokemon_df['Speed'] >= 80].value_counts().sum() cannot be used

value_counts() is typically used to count the occurrences of unique values in a Series.
When applied to a DataFrame, it counts the occurrences of _unique_ rows.
However, using it here might not be the best approach for just counting the number of rows that meet the condition.

The value_counts() method counts unique rows, so duplicates will only be counted once.
On the other hand, shape[0] counts all rows, including duplicates.
'''
filtered_speed_pokemon = pokemon_df[pokemon_df['Speed'] >= 80]
filtered_speed_pokemon.shape[0]
```

```python
#Display this DataFrame using your preferred visualization method.

#hist expects a single column of numerical data to plot a histogram

plt.hist(filtered_speed_pokemon['Speed'], bins=15, color='FireBrick', edgecolor='black')
plt.xlabel('Speed')
plt.ylabel('Frequency')
plt.show()
```

```python
#Find Pokémon with the longest name (excluding spaces)? What is this pokemons name?

longest_pokemon_name_index = pokemon_df['Name'].str.replace(' ', '').str.len().idxmax()
#pokemon_df['Name'].str.replace(' ', '').str.len() - removes spaces from each name and calculates the length of the resulting names.
#.idxmax() - find index of longest name

print(f'The longest pokemon name is {pokemon_df.loc[longest_pokemon_name_index, "Name"]}')
#.loc[]: this method allows to access a group of rows and columns by labels (row indices and column names).
# "Name" - specifies the column from which you want to retrieve data.
```
