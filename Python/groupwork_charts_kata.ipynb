Task 1: Create a bar chart that shows the count of customers for each unique
value in the 'Gender' column (including NaN values). - easy



```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('/content/transaction_dataset.csv')

gender_count = dataset['Gender'].value_counts(dropna=False)

print(gender_count) #just for trying, not really needed

gender_label = gender_count.index.astype(str) # got an error without so seeked how to convert

plt.bar(gender_label, gender_count.values, color = 'blue')
plt.title('Chart for gender')
plt.xlabel('Gender')
plt.ylabel('Amount')

plt.show()
```

Task 2: Create a horizontal bar chart that shows the top 5 most frequent names in the DataFrame, based on the 'name' column. (First, create a grouped DataFrame (name_df), then filter it using iloc, and finally create the visualization.) -medium

```python
name_df = dataset['Name'].value_counts()

top_5 = name_df.iloc[:5]

print(top_5) # just for trying, not really needed

plt.barh(top_5.index, top_5.values, color = 'skyblue')
plt.xlabel('Number')
plt.ylabel('Names')
plt.gca().invert_yaxis() # to have highest value on top
plt.show()
```
